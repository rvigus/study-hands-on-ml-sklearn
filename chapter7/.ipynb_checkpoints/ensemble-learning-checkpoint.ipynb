## Ensemble Learning and Random Forests

Using the outputs from many models to get a final prediction. (Think wisdom of the crowd).

### Voting classifiers

Assume you have trained a number of classifiers on some training data. Individually you get acuracy of about 80% on each. 

<img src="imgs/vc_1.jpg" width="800" height="400">

A very simple way to increase the performance and create an even better classifier is to aggregate the results from the individual models, and predict the class that gets the most votes. **This majority vote classifier is called a hard-voting classifier**.

<img src="imgs/vc_2.jpg" width="800" height="400">

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons
from sklearn.metrics import accuracy_score
```

```python
X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
log_clf = LogisticRegression(random_state=42)
rnd_clf = RandomForestClassifier(random_state=42)
svm_clf = SVC(random_state=42)

voting_clf = VotingClassifier(
    estimators=[("lr", log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard'
)

voting_clf.fit(X_train, y_train);
```

```python
for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

If all models are able to estimate probabilities (i.e have a predict_proba() method), then we can tell sklearn to predict the class with the highest probability, averaged over all individual models.

**This is called soft-voting**. **Generall soft-voting outperforms hard-voting as soft-voting gives more weight to highly confident votes**.

### Bagging and Pasting

One way to get a diverse set of classifiers is to use very different training algorithms as just discussed. Another approach is to use the same training algorithm for every predictor and then train them on different random subsets of the training set. 

When sampling is performed **with replacement** then this method is called **bagging**.

When sampling is performed **without replacement** then this method is called **pasting**.

One of the advantages of such an approach, each of the models can be trained in parralel, via different CPU cores or on different servers.

<img src="imgs/vc_3.jpg" width="800" height="400">

How would we implement such a thing in sklearn?

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500, max_samples=100, bootstrap=True, n_jobs=-1) # bagging because boostrap=True
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)
```

```python
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train);
```

```python
def plot_decision_boundary(mod, X, y, ax):
    
    # Simulate new X features that will cover plot.
    x_min, x_max = X[:, 0].min(), X[:, 0].max()
    y_min, y_max = X[:, 1].min(), X[:, 1].max()
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    Z = mod.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = ax.contour(xx, yy, Z, colors=["k"], linestyles=["--"], linewidths=[0.5])
    ax.set_ylim((y_min, y_max))
    ax.set_xlim((x_min, x_max))

# make plots 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(X_train[:, 0], X_train[:, 1], hue=y_train.ravel(), ax=ax1);
plot_decision_boundary(tree_clf, X_train, y_train, ax1)
ax1.set_title("Single decision tree")

sns.scatterplot(X_train[:, 0], X_train[:, 1], hue=y_train.ravel(), ax=ax2);
plot_decision_boundary(bag_clf, X_train, y_train, ax2)
ax2.set_title("Bagged decision trees")
```

The bagged trees tend to generalize a lot better than the single tree. We can see that the bagged model has less variance than the single decision tree (decision boundary is less irregular). This is because bootstrapping has introduced more diversity into each of the bagged subsets, which means that each individual model will have more bias (sampling a population will not fully fit the specifics of that population), which when averaged creates a more stable result. 

Because of the way sampling works when bootstap=True, it can be that some training samples are never seen by any individual model. These are called *out-of-bag* instances, and we can use these as an additional test set if we like, as a way of giving a first indicator of model performance. See below.

```python
bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500, bootstrap=True, n_jobs=-1, oob_score=True)
bag_clf.fit(X_train, y_train)
print(bag_clf.oob_score_)
```

```python
print(accuracy_score(y_test, bag_clf.predict(X_test)))
```

## Random Forests

Is an esemble of decision trees, generally trained by the bagging method. Where a single decision tree searches for the best feature on which to make a split, Random Forests introduce extra randomness (via sampling) it searches for the best feature amongst a random subset of features. **The algorithm results in greater tree diversity which trades higher bias for lower variance generally resulting in a more generalizable model.**

The final prediction is the most common prediction from each of the individually fit models.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_clf = RandomForestClassifier(n_estimators=500)
rand_clf.fit(X_train, y_train)

rand_clf_2 = RandomForestClassifier(n_estimators=500, max_leaf_nodes=6)
rand_clf_2.fit(X_train, y_train)

# make plots 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
sns.scatterplot(X_train[:, 0], X_train[:, 1], hue=y_train.ravel(), ax=ax1);
plot_decision_boundary(rand_clf, X_train, y_train, ax1)
ax1.set_title("No regularization")

sns.scatterplot(X_train[:, 0], X_train[:, 1], hue=y_train.ravel(), ax=ax2);
plot_decision_boundary(rand_clf_2, X_train, y_train, ax2)
ax2.set_title("With regularization")
```

**Feature importance**

Another nice feature of random forests is that you can do feature selection with them. Feature importance is measured by looking at how much the tree nodes that use that feature reduce impurity on average, across all trees in the forest. We can access that by using the .feature_importances attribute of a fitted model. It is a quick way of determing which features are actually adding value to your model. 

## Boosting

Refers to any ensemble method that combines weak learners into a strong learner. (Remember weak learner is just the term uses for a single model in an ensemble). The general idea of most boosting methods is to train predictors sequentially, each trying to correct the error of its predecessor. 

One of the most popular and successful boosting models is **gradient boosting.**

### Gradient boosting

Works by sequentially adding predictors to an ensemble, each one correcting its predecessor. Each new model added tries to predict the residual error from the chain of models that came before it.

We can demonstrate this below:

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
# new continuous data
x_new = np.linspace(X.min(), X.max(), len(X)).reshape(-1, 1)
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.scatter(X, y)

# 1
tree_reg1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg1.fit(X, y)
y_pred = sum(mod.predict(x_new) for mod in [tree_reg1])
ax.plot(x_new, y_pred, color='r', label='iteration 1')

# 2
y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg2.fit(X, y2)
y_pred = sum(mod.predict(x_new) for mod in [tree_reg1, tree_reg2])
ax.plot(x_new, y_pred, color='g', label='iteration 2')

# 3
y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg3.fit(X, y3)
y_pred = sum(mod.predict(x_new) for mod in [tree_reg1, tree_reg2, tree_reg3])
ax.plot(x_new, y_pred, color='m', label='iteration 3')

ax.legend()
```

```python
# new continuous data
x_new = np.linspace(X.min(), X.max(), len(X)).reshape(-1, 1)
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.scatter(X, y)

y_iter = y.copy()
models = []
for i in range(50):
    
    tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
    tree_reg.fit(X, y_iter)
    y_iter = y_iter - tree_reg.predict(X)
    models.append(tree_reg)
    y_pred = sum(mod.predict(x_new) for mod in models)
    
    if i == 0:
        ax.plot(x_new, y_pred, color='r', label='iteration 1')
    if i == 5:
        ax.plot(x_new, y_pred, color='g', ls='--', label='iteration 5')
    if i == 49:
        ax.plot(x_new, y_pred, color='m', ls='--', label='iteration 50')

ax.legend();
```

As we increase the number of weak learners, we get a closer and closer fit. (Overfitting of course). 

We can achieve the same result using GradientBoosting class from sklearn.

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr_mod = GradientBoostingRegressor(max_depth=2, n_estimators=50, learning_rate=1.0)
gbr_mod.fit(X, y)
preds = gbr_mod.predict(x_new)
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y)
ax.plot(x_new, preds, color='m');
```

For regularization we have all the parameters that you could use on a decision tree. We have the additional parameter in GBR called learning_rate. 

Learning rate scales the contribution of each tree by a constant. The lower the learning rate, the higher the number of models you will need to fit the ensemble, but generally speaking the models generalize better.

We can find the optimal number of trees to use in GBR we can use what is called **early stopping**. Alternatively, you can apply gridsearch over different parameter combinations.

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Get train/val set
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=49)

# Fit model with high number of estimators, which we will then iterate over using staged_predict method
gbr_mod = GradientBoostingRegressor(max_depth=2, n_estimators=120, random_state=42)
gbr_mod.fit(X_train, y_train)
errors = [mean_squared_error(y_val, y_pred) for y_pred in gbr_mod.staged_predict(X_val)]

# of all errors which i gives us lowest.
best_n_estimators = np.argmin(errors) + 1

# Best estimator
gbr_best = GradientBoostingRegressor(max_depth=2, n_estimators=best_n_estimators, random_state=42)
gbr_best.fit(X_train, y_train)

min_error = np.min(errors)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(errors, ls='--')
ax1.plot(best_n_estimators, min_error, "ko")
ax1.plot([0, best_n_estimators], [min_error, min_error], color='k', ls='--')
ax1.plot([best_n_estimators, best_n_estimators], [0, min_error], color='k', ls='--')
ax1.set_ylim(0, max(errors))
ax1.set_xlim(0, 120)
ax1.set_title("Validation error by estimator")

preds = gbr_best.predict(x_new)
ax2.scatter(X, y)
ax2.plot(x_new, preds, color='m')
ax2.set_title(f"GBR with {best_n_estimators}  estimators")
```

Grid search is nice if you want to test many different parameters at once (we can pass more params into the parameter grid).

```python
from sklearn.model_selection import GridSearchCV
```

```python
gbr_mod = GradientBoostingRegressor(max_depth=2)

n_estimators = np.arange(0, 130, step=10)[1:]
print(n_estimators)

grid = GridSearchCV(estimator=gbr_mod, param_grid={'n_estimators':n_estimators}, n_jobs=-1)
grid.fit(X, y)

preds = grid.best_estimator_.predict(x_new)
fig, ax = plt.subplots(1, 1)
ax.scatter(X, y)
ax.plot(x_new, preds, color='m', ls='--');
ax.set_title(f"Best estimators {grid.best_estimator_.n_estimators_}");
```
